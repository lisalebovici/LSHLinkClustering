```python
import numpy as np
import lshlink as lsh
from scipy.cluster.hierarchy import linkage, dendrogram, cophenet
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris().data * 10
```

```python
z = linkage(iris, method="centroid")
```

```python
dendrogram(z)
plt.gcf().set_size_inches(12, 6)
plt.title('Centroid Hierarchical Clustering')
plt.show();
```

```python
C1 = cophenet(z)
```

```python
clusters2, Z2 = lsh.LSHLink(iris, A = 1.4, l = 10, k = 100, dendrogram = True, seed1 = 12, seed2 = 6)
```

```python
C2 = cophenet(Z2)
```

```python
print(np.corrcoef(C1, C2)[0,1])
```
